## StockCode and Description Columns

```python
#check stockcode record
customers['StockCode'].unique()
```

```python
#check number of Description
customers['Description'].unique()
```

```python
#randomly check stockcode number
#show inconsistent writing
customers[customers['StockCode'] == '90014C']
```

```python
#show inconsistent writing
customers[customers['StockCode'] == 21112].sample(10)
```

Indicate that Description was writing inconsistency which is same meaning difference writing so we will clean it to the same text

```python
# clean text
customers['Description'] = customers['Description'].str.lower().str.strip()

# fix inconsistent descriptions
#keep only the most writing style in ach same meaning
desc_map = (
    customers.groupby(['StockCode', 'Description'])
    .size()
    .reset_index(name='count')
    .sort_values(['StockCode','count'], ascending=[True,False])
    .drop_duplicates('StockCode')
)

desc_map

desc_dict = dict(zip(desc_map['StockCode'], desc_map['Description']))

#replace the description in dataframe into consistency
customers['Description'] = customers['StockCode'].map(desc_dict)
```

```python
print("Unique StockCodes:", customers['StockCode'].nunique())
print("Unique Descriptions:", customers['Description'].nunique())
print("Missing Descriptions:", customers['Description'].isna().sum())
```

## InvoiceNo, CustomerID and InvoiceDate Type

```python
#convert datatypes
customers['InvoiceNo'] = customers['InvoiceNo'].astype(int)
customers['CustomerID'] = customers['CustomerID'].astype(int)
customers['InvoiceDate'] = pd.to_datetime(customers['InvoiceDate'])

#check datatype of the columns
customers.dtypes
```

```python
#check format of InvoiceDate
customers['InvoiceDate'].unique()
```

As we check that invoice transection we realize that it shouldn't because of human error, but it could be the promotion or free gift instead so we will drop the row that contain UnitPrice = 0 as well to not confuse and make it wrong for calculate Monterey

## Create TotalSpend Column

```python
#create a new column call TotalSpend
customers['TotalSpend'] = (customers['Quantity'] * customers['UnitPrice'])
```
